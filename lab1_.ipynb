```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
data = {
    'outlook': ['sunny', 'rainy', 'rainy', 'sunny', 'sunny', 'overcast', 'sunny', 'overcast', 'overcast', 'rainy', 'rainy', 'sunny', 'rainy', 'overcast'],
    'temperature': [80.0, 65.0, 71.0, 85.0, 72.0, 64.0, 75.0, 72.0, 83.0, 70.0, 68.0, 69.0, 75.0, 81.0],
    'humidity': [90.0, 70.0, 91.0, 85.0, 95.0, 65.0, 70.0, 90.0, 86.0, 96.0, 80.0, 70.0, 80.0, 75.0],
    'windy': [True, True, True, False, False, True, True, True, False, False, False, False, False, False],
    'play': ['no', 'no', 'no', 'no', 'no', 'yes', 'yes', 'yes', 'yes', 'yes', 'yes', 'yes', 'no', 'yes']
}
```

```python
weather_df = pd.DataFrame(data)
```

```python
for index, row in weather_df.iterrows():
    if row['play'] == 'no':
        plt.scatter(row['temperature'], row['humidity'], s=120, marker='_', linewidths=2, color='red')
    else:
        plt.scatter(row['temperature'], row['humidity'], s=120, marker='+', linewidths=2, color='green')

# Outside the loop to avoid multiple legends
plt.xlabel('Temperature')
plt.ylabel('Humidity')
plt.legend()
plt.show()
```

```python
import numpy as np

def perceptron_algo(X, Y):
    w = np.zeros(len(X[0]))
    eta = 1
    epochs = 10

    for epoch in range(epochs):
        for i, x in enumerate(X):
            if (np.dot(X[i], w) * Y[i]) <= 0:
                w = w + eta * X[i] * Y[i]
    return w
```

```python
X_weather = weather_df[['temperature', 'humidity', 'windy']].values
y_weather = weather_df['play'].map({'no': -1, 'yes': 1}).values

# Call the perceptron algorithm
w = perceptron_algo(X_weather, y_weather)
print("Weight vector:", w)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def perceptron_algo_plot(X, Y):
    '''
    Train perceptron and plot the total loss in each epoch.

    :param X: data samples
    :param Y: data labels
    :return: weight vector as a numpy array
    '''
    w = np.zeros(len(X[0]))
    eta = 1
    n = 30
    errors = []

    for epoch in range(n):
        total_error = 0
        for i, x in enumerate(X):
            if (np.dot(X[i], w) * Y[i]) <= 0:
                total_error += (np.dot(X[i], w) * Y[i])
                w = w + eta * X[i] * Y[i]
        errors.append(total_error * -1)

    plt.plot(errors)
    plt.xlabel('Epoch')
    plt.ylabel('Total Loss')
    plt.title('Perceptron Algorithm - Total Loss over Epochs')
    plt.show()

    return w
```

```python
w = perceptron_algo_plot(X_weather, y_weather)
print("Weight vector:", w)
```
