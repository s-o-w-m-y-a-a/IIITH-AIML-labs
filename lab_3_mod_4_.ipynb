```python
def cal_cost(y_pred, X, y):
    '''
    Calculates the cost for given X and Y.
    y_pred = Vector of y_preds
    X = Row of X's np.zeros((2, j))
    y = Actual y's np.zeros((2, 1))

    where:
        j is the no of features
    '''

    m = len(y)

    predictions = X.dot(y_pred)
    cost = (1 / 2 * m) * np.sum(np.square(predictions - y))

    return cost
```

```python
def minibatch_gradient_descent(X, y, y_pred, learning_rate=0.01, iterations=10, batch_size=20):
    '''
    X = Matrix of X without added bias units
    y = Vector of Y
    y_pred = Vector of y_preds np.random.randn(j, 1)
    learning_rate
    iterations = no of iterations

    Returns the final theta vector and array of cost history over no of iterations
    '''

    m = len(y)
    cost_history = np.zeros(iterations)
    n_batches = int(m / batch_size)

    for it in range(iterations):
        cost = 0.0
        indices = np.random.permutation(m)
        X = X[indices]
        y = y[indices]

        for i in range(0, m, batch_size):
            X_i = X[i: i + batch_size]
            y_i = y[i: i + batch_size]

            X_i = np.c_[np.ones(len(X_i)), X_i]
            prediction = np.dot(X_i, y_pred)

            y_pred = y_pred - (1 / m) * learning_rate * (X_i.T.dot((prediction - y_i)))
            cost += cal_cost(y_pred, X_i, y_i)

        cost_history[it]  = cost

    return y_pred, cost_history
```

```python
lr = 0.1
n_iter = 200
y_pred = np.random.randn(2,1)
y_pred, cost_history = minibatch_gradient_descent(X, y, y_pred, lr, n_iter)

print('y_pred[0]: {:0.3f}\ny_pred[1]: {:0.3f}'.format(y_pred[0][0], y_pred[1][0]))
print('Final error: {:0.3f}'.format(cost_history[-1]))
```

2.Test all the three out on real datasets.

Applying batch, stochastic, and mini-batch gradient descent on a real dataset using Python's scikit-learn library.

```python
from sklearn.datasets import load_diabetes
from sklearn.preprocessing import StandardScaler
import numpy as np
import matplotlib.pyplot as plt

# Load a real dataset
diabetes = load_diabetes()
X = diabetes.data
y = diabetes.target.reshape(-1, 1)

# Data preprocessing - scaling features
scaler = StandardScaler()
X = scaler.fit_transform(X)

# Cost function
def cal_cost(y_pred, X, y):
    m = len(y)
    predictions = X.dot(y_pred)
    cost = (1 / (2 * m)) * np.sum(np.square(predictions - y))
    return cost

# Gradient Descent Functions
def batch_gradient_descent(X, y, y_pred, learning_rate=0.01, iterations=100):
    m = len(y)
    cost_history = np.zeros(iterations)

    for it in range(iterations):
        prediction = np.dot(X, y_pred)
        y_pred = y_pred - (1 / m) * learning_rate * (X.T.dot((prediction - y)))
        cost_history[it] = cal_cost(y_pred, X, y)

    return y_pred, cost_history

def stochastic_gradient_descent(X, y, y_pred, learning_rate=0.01, iterations=100):
    m = len(y)
    cost_history = np.zeros(iterations)

    for it in range(iterations):
        cost = 0.0

        for i in range(m):
            rand_ind = np.random.randint(0, m)
            X_i = X[rand_ind, :].reshape(1, -1)
            y_i = y[rand_ind].reshape(1, 1)
            prediction = np.dot(X_i, y_pred)

            y_pred = y_pred - (1 / m) * learning_rate * (X_i.T.dot((prediction - y_i)))
            cost += cal_cost(y_pred, X_i, y_i)

        cost_history[it] = cost

    return y_pred, cost_history

def minibatch_gradient_descent(X, y, y_pred, learning_rate=0.01, iterations=100, batch_size=20):
    m = len(y)
    cost_history = np.zeros(iterations)
    n_batches = int(m / batch_size)

    for it in range(iterations):
        cost = 0.0
        indices = np.random.permutation(m)
        X = X[indices]
        y = y[indices]

        for i in range(0, m, batch_size):
            X_i = X[i: i + batch_size]
            y_i = y[i: i + batch_size]

            prediction = np.dot(X_i, y_pred)
            y_pred = y_pred - (1 / m) * learning_rate * (X_i.T.dot((prediction - y_i)))
            cost += cal_cost(y_pred, X_i, y_i)

        cost_history[it] = cost

    return y_pred, cost_history

# Initialize parameters
lr = 0.1
n_iter = 200
batch_size = 20

# Initialize y_pred
y_pred = np.random.randn(X.shape[1], 1)

# Run all three gradient descent methods
y_pred_batch, cost_history_batch = batch_gradient_descent(X, y, y_pred, lr, n_iter)
y_pred_stochastic, cost_history_stochastic = stochastic_gradient_descent(X, y, y_pred, lr, n_iter)
y_pred_minibatch, cost_history_minibatch = minibatch_gradient_descent(X, y, y_pred, lr, n_iter, batch_size)

# Plot the cost history for all methods
plt.figure(figsize=(10, 6))
plt.plot(range(n_iter), cost_history_batch, label='Batch GD')
plt.plot(range(n_iter), cost_history_stochastic, label='Stochastic GD')
plt.plot(range(n_iter), cost_history_minibatch, label='Mini-batch GD')
plt.xlabel('Number of iterations')
plt.ylabel('Error')
plt.legend()
plt.title('Error vs. Number of iterations for different GD methods')
plt.show()
```

This code uses the diabetes dataset from scikit-learn, applies feature scaling, and runs batch, stochastic, and mini-batch gradient descent on the dataset. It then plots the error (cost) against the number of iterations for comparison. Adjust the parameters and algorithms as needed for your analysis.

3.Compare the effects of changing learning rate by the same amount in Batch GD, SGD and Mini-batch GD.


comparing the effects of changing the learning rate for batch, stochastic, and mini-batch gradient descent using scikit-learn's diabetes dataset

```python
from sklearn.datasets import load_diabetes
from sklearn.preprocessing import StandardScaler
import numpy as np
import matplotlib.pyplot as plt

# Load the diabetes dataset
diabetes = load_diabetes()
X = diabetes.data
y = diabetes.target.reshape(-1, 1)

# Feature scaling
scaler = StandardScaler()
X = scaler.fit_transform(X)

# Cost function
def cal_cost(y_pred, X, y):
    m = len(y)
    predictions = X.dot(y_pred)
    cost = (1 / (2 * m)) * np.sum(np.square(predictions - y))
    return cost

# Gradient Descent Functions
def batch_gradient_descent(X, y, y_pred, learning_rate=0.01, iterations=100):
    m = len(y)
    cost_history = np.zeros(iterations)

    for it in range(iterations):
        prediction = np.dot(X, y_pred)
        y_pred = y_pred - (1 / m) * learning_rate * (X.T.dot((prediction - y)))
        cost_history[it] = cal_cost(y_pred, X, y)

    return y_pred, cost_history

def stochastic_gradient_descent(X, y, y_pred, learning_rate=0.01, iterations=100):
    m = len(y)
    cost_history = np.zeros(iterations)

    for it in range(iterations):
        cost = 0.0

        for i in range(m):
            rand_ind = np.random.randint(0, m)
            X_i = X[rand_ind, :].reshape(1, -1)
            y_i = y[rand_ind].reshape(1, 1)
            prediction = np.dot(X_i, y_pred)

            y_pred = y_pred - (1 / m) * learning_rate * (X_i.T.dot((prediction - y_i)))
            cost += cal_cost(y_pred, X_i, y_i)

        cost_history[it] = cost

    return y_pred, cost_history

def minibatch_gradient_descent(X, y, y_pred, learning_rate=0.01, iterations=100, batch_size=20):
    m = len(y)
    cost_history = np.zeros(iterations)
    n_batches = int(m / batch_size)

    for it in range(iterations):
        cost = 0.0
        indices = np.random.permutation(m)
        X = X[indices]
        y = y[indices]

        for i in range(0, m, batch_size):
            X_i = X[i: i + batch_size]
            y_i = y[i: i + batch_size]

            prediction = np.dot(X_i, y_pred)
            y_pred = y_pred - (1 / m) * learning_rate * (X_i.T.dot((prediction - y_i)))
            cost += cal_cost(y_pred, X_i, y_i)

        cost_history[it] = cost

    return y_pred, cost_history

# Initialize parameters
n_iter = 100
learning_rates = [0.001, 0.01, 0.1]  # Different learning rates to compare

# Initialize y_pred
y_pred = np.random.randn(X.shape[1], 1)

# Run gradient descent for different learning rates
for lr in learning_rates:
    y_pred_batch, cost_history_batch = batch_gradient_descent(X, y, y_pred.copy(), lr, n_iter)
    y_pred_stochastic, cost_history_stochastic = stochastic_gradient_descent(X, y, y_pred.copy(), lr, n_iter)
    y_pred_minibatch, cost_history_minibatch = minibatch_gradient_descent(X, y, y_pred.copy(), lr, n_iter, batch_size=20)

    # Plot the cost history for each learning rate
    plt.figure(figsize=(10, 6))
    plt.plot(range(n_iter), cost_history_batch, label=f'Batch GD - LR: {lr}')
    plt.plot(range(n_iter), cost_history_stochastic, label=f'Stochastic GD - LR: {lr}')
    plt.plot(range(n_iter), cost_history_minibatch, label=f'Mini-batch GD - LR: {lr}')
    plt.xlabel('Number of iterations')
    plt.ylabel('Error')
    plt.legend()
    plt.title(f'Error vs. Number of iterations for different GD methods (LR: {lr})')
    plt.show()
```
